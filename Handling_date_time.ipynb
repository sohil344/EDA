```python
import pandas as pd
import numpy as np
```

**Working with Date Time**

```python
#converting to datetime datatype
date['date']=pd.to_datetime(date['date'])
```

```python
date.info()
```

```python
#Extract Year
date['date_year']=date['date'].dt.year
date.sample(10)
```

```python
#Extract Month

date['date_month']=date['date'].dt.month
date.sample(5)
```

```python
#Extract Month name
date['date_month_name']=date['date'].dt.month_name()
date.sample(5
            )
```

```python
#Extract day from date

date['date_day']=date['date'].dt.day
date.sample(5)
```

```python
#Extract day num in week
date['date_dow']=date['date'].dt.dayofweek
date.sample(5)
```

```python
# day of week - name
date['date_dow_name']=date['date'].dt.day_name()
date=date.drop(columns=['product_id','city_id'])
```

```python
date.sample(5)
```

```python
# is weekend?

date['date_weekend']=np.where(date['date_dow_name'].isin(['Saturday','Sunday']),1,0)
date.sample(5)
```

```python
date=date.drop(columns=['orders'])
```

```python
#Extract week of year
date['date_week']=date['date'].dt.week
date.sample(5)
```

```python
#Extract time elapsed between dates

import datetime
today=datetime.datetime.today()
today
```

```python
today-date['date']
```

```python
#Elapsed days 
(today-date['date']).dt.days
```

```python
time['date']=pd.to_datetime(time['date'])
```

```python
time['date']
```

```python
time['hour']=time['date'].dt.hour
time['min']=time['date'].dt.minute
time['sec']=time['date'].dt.second
```

```python
time.head()
```

```python
time['time']=time['date'].dt.time
time.sample(5)
```

```python
today-time['date']
```

```python
# in hours

time['diff_hrs']=(today - time['date'])/np.timedelta64(1,'h')

# in hours

time['diff_min']=(today - time['date'])/np.timedelta64(1,'m')

# in seconds

time['diff_sec']=(today - time['date'])/np.timedelta64(1,'s')

```
