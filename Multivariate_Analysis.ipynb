```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris=sns.load_dataset('iris')
titanic=pd.read_csv('/content/titanic_train.csv')
flights=sns.load_dataset('flights')
tips=sns.load_dataset('tips')
```

**Scatter Plot**

Numerical-Numerical datatype

```python
sns.scatterplot(tips['total_bill'],tips['tip'],hue=tips['sex'],style=tips['smoker'],size=tips['size'])
```

```python
titanic.head()
```

**Bar-Plot**

Numerical-Categorical datatype

```python
sns.barplot(titanic['pclass'],titanic['age'],hue=titanic['sex'])
```

**Box Plot (Numerical - Categorical)**

```python
sns.boxplot(titanic['sex'],titanic['age'],hue=titanic['survived'])
```

**Distplot (Numerical-Categorical)**

```python
sns.distplot(titanic[titanic['survived']==0]['age'],hist=False)#blue
sns.distplot(titanic[titanic['survived']==1]['age'],hist=False)#orange
```

**HeatMap---(Categorical & Categorical)**

```python
sns.heatmap(pd.crosstab(titanic['pclass'],titanic['survived']))
```

```python
pd.crosstab(titanic['pclass'],titanic['survived'])
```

```python
 #percentage of survival rate in each class
titanic.groupby('pclass').mean()['survived']*100 
```

```python
#percentage of survival rate in sex category
titanic.groupby('sex').mean()['survived']*100 
```

**PairPlot**

```python
sns.pairplot(iris,hue='species')
```
